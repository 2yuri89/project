```python
import pandas as pd
```

```python
# 데이터 불러오기
df = pd.read_csv('C:/Users/human/2nd_project/eco/last/Data_Preprocessing/merged_aptinfo_distance_politics_transactionCount_E_pop_density_df.csv')
service = pd.read_csv('C:/Users/human/2nd_project/eco/last/Data_Preprocessing/서울시 상권분석서비스(소득소비-행정동).csv',encoding = 'cp949')
```

```python
# nan값 제거

df['contract_date'].isna().sum()
df[df['contract_date'].isnull()]

# 중복제거
df = df.drop(index=18084)
# df[df['gu_adong']=='강동구_둔촌2동']

# 분기추출
df['contract_quarter_code'] = (df['contract_date'].str[:4]  # 연도 추출
    + (df['contract_date'].str[5:7].astype(int) // 4 + 1).astype(str) 
)
# 분기별 분리
df['contract_quarter_code'] = df['contract_quarter_code'].astype('int64')

df.info()
```

```python
# 2024 년 제거
service_filtered = service[~service['기준_년분기_코드'].astype(str).str.contains('2024')]
service_filtered
```

```python
# 원본파일에는 있지만 소득파일에는 없는 동 확인
for c in df['a_dong'].unique():
    if c not in service['행정동_코드_명'].unique():
        print(c)
```

```python
# 행정동코드명 이름 바꾸기
service_filtered['행정동_코드_명'] = service_filtered['행정동_코드_명'].str.replace('?','·')
service_filtered['행정동_코드_명'] = service_filtered['행정동_코드_명'].str.replace('일원2동','개포3동')
```

```python
# 상일동 분리
service_filtered1 = service_filtered[service_filtered['행정동_코드_명'] == '상일동'].copy()
service_filtered1['행정동_코드_명'] = '상일1동'  # 상일1동으로 수정

service_filtered2 = service_filtered[service_filtered['행정동_코드_명'] == '상일동'].copy()
service_filtered2['행정동_코드_명'] = '상일2동'

service_filtered_result = pd.concat([service_filtered, service_filtered1, service_filtered2], ignore_index=True)
```

```python
# 결측치 제거
def impute_zero(row):
    for c in row.index[5:]:
        if row[c] == 0:
            q = str(row['기준_년분기_코드'])[-1]            
            q_mask = service_filtered_result['기준_년분기_코드'].map(lambda x: (str(x)[-1] == q))
            dn = row['행정동_코드_명']
            dn_mask = service_filtered_result['행정동_코드_명'].map(lambda x: x == row['행정동_코드_명'])
            mean_v = service_filtered_result[q_mask & dn_mask][c].mean()
            if mean_v == 0:
                mean_v = service_filtered_result[dn_mask][c].mean()

            row[c] = mean_v
    
    return row

service_filtered_result_imputed = service_filtered_result.apply(impute_zero, axis=1)        

service_filtered_result.iloc[:10, :].apply(lambda row: print(row["행정동_코드"]), axis=1)
```

```python
merged_df = df.merge(service_filtered_result_imputed, 
                     left_on=['a_dong', 'contract_quarter_code'], 
                     right_on=['행정동_코드_명', '기준_년분기_코드'], 
                     how='left')
```

```python
# 결측치 처리
def impute_zero2(row):
    if row.name % 100 == 0:
        print(row.name, "번째 데이터를 처리합니다.")

    for c in row.index[54:]:
        if row[c] == 0:
            q = str(row['기준_년분기_코드'])[-1]            
            q_mask = merged_df['기준_년분기_코드'].map(lambda x: (str(x)[-1] == q))
            dn = row['행정동_코드_명']
            dn_mask = merged_df['행정동_코드_명'].map(lambda x: x == row['행정동_코드_명'])
            mean_v = merged_df[q_mask & dn_mask][c].mean()
            if mean_v == 0:
                mean_v = merged_df[dn_mask][c].mean()
                if mean_v == 0:
                    gu_mask = (merged_df["gu"] == row['gu'])
                    mean_v = merged_df[gu_mask][c].mean()

            row[c] = mean_v
    
    return row

# merged_df_imputed = merged_df.apply(impute_zero2, axis=1)        

# service_filtered_result.iloc[:10, :].apply(lambda row: print(row["행정동_코드"]), axis=1)
```

```python
# 결측치 처리 함수 돌리기
copied_merged_df = merged_df.copy()
mask = copied_merged_df.iloc[:, 54:].isin([0]).sum(axis=1) > 0
include_zero_df = copied_merged_df[mask]

merged_df_imputed2 = include_zero_df.apply(impute_zero2, axis=1) 
```
