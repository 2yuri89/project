```python
import pandas as pd

def process_apt_df(df):
    # 시구, 동 컬럼 생성
    df['city_gu'] = df['시군구'].str.split('구 ').str[0] + '구'
    df['dong'] = df['시군구'].str.split('구 ').str[1]

    # 기존주소와 도로명주소 생성
    df['기존주소'] = df['시군구'] + ' ' + df['번지']
    df['도로명주소'] = df['city_gu'] + ' ' + df['도로명']

    # 사용할 컬럼만 추출
    df = df[['NO', '기존주소', '도로명주소']]
    
    return df
```

```python
dataframes = [apt2013_df, apt2014_df, apt2015_df, apt2016_df, apt2017_df, 
              apt2018_df, apt2019_df, apt2020_df, apt2021_df, apt2022_df, 
              apt2023_df, apt2024_df]

processed_dataframes = [process_apt_df(df) for df in dataframes]
# 이후 processed_dataframes는 각 연도별로 처리된 데이터프레임들의 리스트가 됩니다.

apt2013_df, apt2014_df, apt2015_df, apt2016_df, apt2017_df, apt2018_df, apt2019_df, apt2020_df, apt2021_df, apt2022_df, apt2023_df, apt2024_df = processed_dataframes
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 준비
years = ['2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023', '2024']
transactions = [
    len(apt2013_df), len(apt2014_df), len(apt2015_df), len(apt2016_df), 
    len(apt2017_df), len(apt2018_df), len(apt2019_df), len(apt2020_df), 
    len(apt2021_df), len(apt2022_df), len(apt2023_df), len(apt2024_df)
]

# 색상 팔레트 준비
colors = ['lightgray']*6 + [
    '#FF6B6B',   # 2019 - 산호색 계열
    '#4ECDC4',   # 2020 - 청록색
    '#45B7D1',   # 2021 - 밝은 파란색
    '#FDCB6E',   # 2022 - 따뜻한 노란색
    '#6C5CE7'    # 2023 - 보라색
] + ['lightgray']  # 2024

# 그래프 설정
plt.figure(figsize=(9, 7))
bars = plt.bar(years, transactions, color=colors)

# 제목과 라벨 설정
plt.title('연도별 총 거래 건수 (2019-2023 하이라이트)', fontsize=16)
plt.xlabel('연도', fontsize=12)
plt.ylabel('총 거래 건수', fontsize=12)

# 각 막대 위에 거래 건수 표시
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height):,}', 
             ha='center', va='bottom', fontsize=10)

# x축 라벨 회전 (가독성을 위해)
plt.xticks(rotation=45)

# 테두리 제거
ax = plt.gca()  # 현재 축 가져오기
ax.spines['top'].set_visible(False)  # 위쪽 테두리 제거
ax.spines['right'].set_visible(False)  # 오른쪽 테두리 제거
ax.spines['left'].set_visible(False)  # 왼쪽 테두리 제거
ax.spines['bottom'].set_visible(False)  # 아래쪽 테두리 제거

# 레이아웃 조정
plt.tight_layout()

# 그래프 저장
plt.savefig("../yearly_transactions_highlighted.png", transparent=True)

# 그래프 표시
plt.show()
```
