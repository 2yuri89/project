```python
import pandas as pd
from sklearn.model_selection import train_test_split



df['year'] = df.index.str.split('_').str[0]


y = df['adjusted_price']  


sampled_df = pd.DataFrame()

for year in range(2019, 2024):

    year_df = df[df['year'] == str(year)]
    

    _, sampled_data = train_test_split(year_df, test_size=3000, random_state=42)
    

     sampled_df = pd.concat([sampled_df, sampled_data])


sampled_df
```

```python
sampled_df['year'] = pd.to_numeric(sampled_df['year'])
```

```python
sampled_non_2023 = sampled_df[sampled_df['year'] !=2023]
sampled_2023_data = sampled_df[sampled_df['year'] ==2023]

print(sampled_non_2023.shape, sampled_2023_data.shape)
```

```python
sampled_non_2023 = sampled_non_2023.drop(columns=['year'])
```

```python
y = sampled_non_2023['adjusted_price']
X = sampled_non_2023.drop(columns=['adjusted_price'])
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
# 하이퍼 파라미터 조정 전 , 100 = 28306
rf_regressor = RandomForestRegressor(random_state=42)

rf_regressor.fit(X_train, y_train)


y_pred = rf_regressor.predict(X_test)


mse = mean_squared_error(y_test, y_pred)
rmse = mse ** 0.5
print(f"RMSE: {rmse:.4f}")
```
